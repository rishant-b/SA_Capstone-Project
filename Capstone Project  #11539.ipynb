```python
df = pd.concat([train_data, test_data])
df
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
 
lenc = LabelEncoder()
df.ad = lenc.fit_transform(df['ad']) 

enc = pd.get_dummies(df.adgroup)
df = pd.concat([df,enc], axis = 1)
```

```python
df["weekday"] = pd.to_datetime(df.date).dt.weekday
df["month"] = pd.to_datetime(df.date).dt.month
df["year"] = pd.to_datetime(df.date).dt.year
```

```python
df.drop(["date","campaign","adgroup"], axis = 1, inplace = True)
```

```python
df.dtypes
```

```python
df.shape
```

```python
X_train = df.iloc[0:4571, :].drop(["revenue"], axis =1)
y_train = df.iloc[0:4571, :].revenue
X_test = df.iloc[4571:, :].drop(["revenue"], axis =1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.linear_model import RidgeCV, LassoCV
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor


## Import evaluation metrics
from sklearn.metrics import r2_score, mean_squared_error
```

```python
lr = LinearRegression()

lr.fit(X_train,y_train)
y_pred_test = lr.predict(X_test)
y_pred_train = lr.predict(X_train)

print("R2_Score for training data:", r2_score(y_train, y_pred_train))
print("RMSE for training data:", mean_squared_error(y_train, y_pred_train))

# Test RMSE = 210.12
```

```python
ls = Lasso()

ls.fit(X_train,y_train)
y_pred_test = ls.predict(X_test)
y_pred_train = ls.predict(X_train)

print("R2_Score for training data:", r2_score(y_train, y_pred_train))
print("RMSE for training data:", mean_squared_error(y_train, y_pred_train))

# Test RMSE = 212.99
```

```python
lsCV = LassoCV()

lsCV.fit(X_train,y_train)
y_pred_test = lsCV.predict(X_test)
y_pred_train = lsCV.predict(X_train)

print("R2_Score for training data:", r2_score(y_train, y_pred_train))
print("RMSE for training data:", mean_squared_error(y_train, y_pred_train))
```

```python
rdg = Ridge()

rdg.fit(X_train,y_train)
y_pred_test = rdg.predict(X_test)
y_pred_train = rdg.predict(X_train)

print("R2_Score for training data:", r2_score(y_train, y_pred_train))
print("RMSE for training data:", mean_squared_error(y_train, y_pred_train))
```

```python
rdgCV = RidgeCV()

rdgCV.fit(X_train,y_train)
y_pred_test = rdgCV.predict(X_test)
y_pred_train = rdgCV.predict(X_train)

print("R2_Score for training data:", r2_score(y_train, y_pred_train))
print("RMSE for training data:", mean_squared_error(y_train, y_pred_train))
```

```python
gbr = GradientBoostingRegressor(n_estimators = 100)

gbr.fit(X_train,y_train)
y_pred_test = gbr.predict(X_test)
y_pred_train = gbr.predict(X_train)

print("R2_Score for training data:", r2_score(y_train, y_pred_train))
print("RMSE for training data:", mean_squared_error(y_train, y_pred_train))

# Test RMSE = 141.10
```

```python
gbr = GradientBoostingRegressor(n_estimators = 500)

gbr.fit(X_train,y_train)
y_pred_test = gbr.predict(X_test)
y_pred_train = gbr.predict(X_train)

print("R2_Score for training data:", r2_score(y_train, y_pred_train))
print("RMSE for training data:", mean_squared_error(y_train, y_pred_train))

# Test RMSE = 143.73
```

```python
adboost = AdaBoostRegressor(n_estimators = 100)

adboost.fit(X_train,y_train)
y_pred_test = adboost.predict(X_test)
y_pred_train = adboost.predict(X_train)

print("R2_Score for training data:", r2_score(y_train, y_pred_train))
print("RMSE for training data:", mean_squared_error(y_train, y_pred_train))

# Test RMSE = 185.68
```

```python
rf = RandomForestRegressor(n_estimators = 500)

rf.fit(X_train,y_train)
y_pred_test = rf.predict(X_test)
y_pred_train = rf.predict(X_train)

print("R2_Score for training data:", r2_score(y_train, y_pred_train))
print("RMSE for training data:", mean_squared_error(y_train, y_pred_train))

# Test RMSE = 157.36
```

```python
dtr = DecisionTreeRegressor(max_depth = 5)

dtr.fit(X_train,y_train)
y_pred_test = dtr.predict(X_test)
y_pred_train = dtr.predict(X_train)

print("R2_Score for training data:", r2_score(y_train, y_pred_train))
print("RMSE for training data:", mean_squared_error(y_train, y_pred_train))

# Test RMSE = 148.86
```

```python
res = pd.DataFrame(y_pred_test)  
res.columns = ["revenue"]
res.to_csv("prediction_results_dtr.csv", index = False)      
```
